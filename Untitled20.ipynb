# Máquinas de Vectores de Soporte (SVM) – Hiperplanos de Separación


## 1. Explicación Teórica y Demostraciones Matemáticas

### Fundamentos de clasificación en SVM

Una **Máquina de Vectores de Soporte (SVM)** es un método de aprendizaje supervisado para clasificación binaria (dos clases). Su objetivo es encontrar un **hiperplano** que separe de forma óptima las muestras de las dos clases, asignándolas a uno u otro lado del hiperplano según su etiqueta. En un espacio $p$-dimensional, un *hiperplano* es un subespacio plano de dimensión $p-1$ que divide el espacio en dos semiespacios. En 2 dimensiones un hiperplano es una línea recta, en 3 dimensiones es un plano común, y en dimensiones superiores es difícil de visualizar pero se define de forma análoga.

**Definiciones clave:** En el contexto de SVM, podemos definir:

- **Hiperplano de separación:** Frontera de decisión lineal entre dos clases, definida por $\mathbf{w}\cdot \mathbf{x} + b = 0$ (donde $\mathbf{w}$ es un vector normal al hiperplano y $b$ es el sesgo o término independiente)
Cómo implementar máquinas de vector soporte desde cero. Los puntos en un lado del hiperplano satisfarán $\mathbf{w}\cdot \mathbf{x} + b > 0$ y se asignarán a una clase, mientras que los del otro lado tendrán $\mathbf{w}\cdot \mathbf{x} + b < 0$ y pertenecerán a la clase opuesta .

- **Vectores de soporte:** Las **muestras más cercanas** al hiperplano, aquellas que “soportan” o determinan la posición óptima del hiperplano. Estos puntos son críticos ya que el hiperplano óptimo **depende únicamente de ellos**; mover un vector de soporte afectaría la solución, mientras que otros puntos más alejados no la alteran.

- **Margen:** La **distancia mínima** desde el hiperplano a las muestras de entrenamiento más cercanas (pertenecientes a cualquier clase). En otras palabras, el margen es la distancia desde el hiperplano a los vectores de soporte. Cuanto más grande sea el margen, más separadas estarán las clases del hiperplano, lo que suele indicar una mejor generalización.

En un problema separable linealmente, **infinitos hiperplanos** podrían separar correctamente las dos clases (ver figura hipotética de abajo). SVM propone un criterio único para elegir *el mejor* hiperplano: seleccionar aquel que **maximiza el margen** de separación entre clases. Este hiperplano recibe el nombre de *hiperplano de margen máximo* (*maximum margin hyperplane*). La intuición es que al maximizar la distancia a los puntos más cercanos de ambas clases, el clasificador tendrá mayor tolerancia al ruido y mejor generalización.

### Derivación matemática de los hiperplanos de separación

Para concretar estas ideas, consideremos un conjunto de entrenamiento con $n$ ejemplos $\{(\mathbf{x}_i, y_i)\}_{i=1}^n$, donde $\mathbf{x}_i \in \mathbb{R}^p$ y $y_i \in \{+1, -1\}$ indica la clase. Un hiperplano lineal se expresa como:

$$
\mathbf{w}\cdot \mathbf{x} + b = 0,
$$

con $\mathbf{w}\in \mathbb{R}^p$ un vector de pesos (normal al hiperplano) y $b\in \mathbb{R}$ el sesgo. Este hiperplano clasifica un punto $\mathbf{x}$ según el signo de $f(\mathbf{x}) = \mathbf{w}\cdot \mathbf{x} + b$. Para que el hiperplano **separe correctamente** todas las muestras, debemos tener:

$$
y_i(\mathbf{w}\cdot \mathbf{x}_i + b) > 0, \qquad \forall i=1,\dots,n,
$$

lo que asegura que cada punto $\mathbf{x}_i$ caiga en el semiespacio positivo si $y_i=+1$ o en el negativo si $y_i=-1$. Sin embargo, esta condición por sí sola no define un único hiperplano (cualquier separación funciona); necesitamos imponer un criterio adicional.

Definamos el **margen geométrico** de un hiperplano como la distancia mínima de cualquier muestra al hiperplano. Matemáticamente, la distancia de un punto $\mathbf{x}_0$ al hiperplano $(\mathbf{w},b)$ viene dada por:

$$
\text{dist}(\mathbf{x}_0) = \frac{|\,\mathbf{w}\cdot \mathbf{x}_0 + b\,|}{\|\mathbf{w}\|},
$$

donde $\|\mathbf{w}\|$ es la norma Euclidea de $\mathbf{w}$ . El **margen $\tau$** del hiperplano respecto al conjunto de datos es la mínima distancia de este tipo considerando todas las muestras de entrenamiento. Para maximizar este margen, es equivalente (por conveniencia matemática) a minimizar $\|\mathbf{w}\|$. Podemos fijar una escala de $\mathbf{w}$ y $b$ aprovechando la homogeneidad: cualquier $(\mathbf{w},b)$ que define el mismo hiperplano puede escalarse arbitrariamente (si $\mathbf{w}\cdot \mathbf{x}+b=0$, entonces $(c\mathbf{w})\cdot \mathbf{x}+ (c b)=0$ también para $c>0$). Por convenio se suele imponer que **los vectores de soporte queden a una distancia margen exacta de 1** del hiperplano, es decir:

$$
\min_{i} y_i(\mathbf{w}\cdot \mathbf{x}_i + b) = 1,
$$

lo que geométricamente equivale a $\tau \|\mathbf{w}\| = 1$ . Con esta normalización, los puntos más cercanos cumplen $y_i(\mathbf{w}\cdot \mathbf{x}_i + b) = 1$, y para todos los puntos se tendrá $y_i(\mathbf{w}\cdot \mathbf{x}_i + b)\ge 1$. En adelante asumiremos que hemos escalado $\mathbf{w},b$ de esta manera, así el margen geométrico es $\tau = 1/\|\mathbf{w}\|$.

Bajo esta convención, podemos formular la búsqueda del hiperplano de margen máximo como un problema de optimización **convexo** (específicamente, un problema de **optimización cuadrática (QP)**) de la siguiente forma:

> **(Problema primal SVM, margen duro):**  
> Minimizar $\displaystyle \frac{1}{2}\|\mathbf{w}\|^2$  
> Sujeto a las restricciones $y_i(\mathbf{w}\cdot \mathbf{x}_i + b) \ge 1$ para $i=1,\dots,n$.

Aquí, minimizar $\frac{1}{2}\|\mathbf{w}\|^2$ (equivalente a maximizar el margen $1/\|\mathbf{w}\|$) está condicionado a que *todas* las muestras estén clasificadas correctamente con margen al menos 1. Este es un problema QP estándar: la función objetivo es cuadrática (convexa) y las restricciones son lineales. Si los datos son linealmente separables, este problema tiene solución y proporciona el **hiperplano separador óptimo** (denominado de *margen duro*). En la práctica, suele resolverse mediante métodos matemáticos de optimización (ver siguiente subsección).

**Nota:** Si existe una solución factible para este problema, será única para $\mathbf{w}$ (salvo escala) y $b$, ya que la función objetivo es estrictamente convexa; de lo contrario (si los datos no son separables sin error), el problema es infeasible y necesitamos relajar las restricciones.

### Márgenes suaves y criterio de optimización con errores

En problemas del mundo real, es frecuente que los datos no sean perfectamente separables linealmente. Puede haber ruido, solapamiento de clases, o incluso clases intrínsecamente no separables en el espacio original. Para lidiar con esto, **Cortes y Vapnik (1995)** introdujeron la idea de permitir ciertos errores o violaciones del margen a cambio de obtener un modelo más robusto. Esta extensión se conoce como **SVM de margen suave (soft margin)**.

En la formulación de margen suave, se introducen variables de holgura $\xi_i \ge 0$ para cada muestra, que miden cuánto **viola** cada punto la condición de margen. La restricción se relaja a:

$$
y_i(\mathbf{w}\cdot \mathbf{x}_i + b) \ge 1 - \xi_i, \qquad \xi_i \ge 0.
$$

Si $\xi_i = 0$, la muestra $i$ cumple el margen normal ($y_i(\mathbf{w}\cdot \mathbf{x}_i+b)\ge 1$); si $\xi_i > 0$, la muestra cae dentro de la “zona de margen” o incluso puede estar mal clasificada si $\xi_i > 1$. Permitir $\xi_i > 0$ hace siempre factible el problema, a costa de penalizar estas violaciones en la función objetivo.

La formulación de optimización de SVM con margen suave es :

> **(Problema primal SVM, margen suave):**  
> Minimizar $\displaystyle \frac{1}{2}\|\mathbf{w}\|^2 + C \sum_{i=1}^n \xi_i$  
> Sujeto a $y_i(\mathbf{w}\cdot \mathbf{x}_i + b) \ge 1 - \xi_i$, con $\xi_i \ge 0$ para $i=1,\dots,n$.

El término $\frac{1}{2}\|\mathbf{w}\|^2$ intenta maximizar el margen, mientras que $C \sum \xi_i$ penaliza la suma de violaciones del margen (si $C$ es grande, las violaciones se penalizan fuertemente, forzando menos errores pero un margen más estrecho; si $C$ es pequeño, se permite más error a cambio de un margen más amplio).


### Introducción al problema de optimización cuadrática asociado

La resolución analítica del problema de optimización anterior se aborda mediante técnicas de optimización convexa. Un enfoque estándar es construir la **función Lagrange** del problema primal (hard margin, para empezar) y obtener su dual.

Para el caso separable sin errores, la función Lagrangiana (introduciendo multiplicadores de Lagrange $\alpha_i \ge 0$ para las restricciones) es:

$$
\mathcal{L}( \mathbf{w}, b, \boldsymbol{\alpha} ) \;=\; \frac{1}{2}\|\mathbf{w}\|^2 \;-\; \sum_{i=1}^n \alpha_i\,[\,y_i(\mathbf{w}\cdot \mathbf{x}_i + b) - 1\,],
$$

donde $\alpha_i \ge 0$. Las condiciones de Karush-Kuhn-Tucker (KKT) necesarias para el óptimo nos dan:

- Derivando respecto a $\mathbf{w}$: $\displaystyle \frac{\partial \mathcal{L}}{\partial \mathbf{w}} = \mathbf{w} - \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i = \mathbf{0}$. De aquí se obtiene la importante condición:

  $$\mathbf{w}^* = \sum_{i=1}^n \alpha_i^*\,y_i\,\mathbf{x}_i,$$

  es decir, el vector normal óptimo $\mathbf{w}^*$ es una **combinación lineal de las muestras de entrenamiento**, ponderadas por los multiplicadores $\alpha_i$. Notemos que en la solución óptima muchos $\alpha_i$ serán cero; de hecho, resulta que *solo* las muestras correspondientes a las restricciones activas (las más cercanas al hiperplano) tienen $\alpha_i > 0$. Estas son precisamente los **vectores de soporte**. Esto confirma formalmente que el hiperplano viene determinado por los vectores de soporte.

- Derivando respecto a $b$: $\displaystyle \frac{\partial \mathcal{L}}{\partial b} = \sum_{i=1}^n \alpha_i y_i = 0$. Esta condición garantiza que en el óptimo la contribución positiva y negativa balancea, cumpliendo $\sum_i \alpha_i y_i = 0$ .

- Condición de complementariedad (KKT): $\alpha_i [\,y_i(\mathbf{w}\cdot \mathbf{x}_i + b) - 1\,] = 0$ para cada $i$ . Esto implica que, para cada muestra $i$, o bien $\alpha_i = 0$ (no es vector de soporte) o bien $y_i(\mathbf{w}\cdot \mathbf{x}_i + b) = 1$ (si $\alpha_i>0$, la muestra está exactamente sobre el margen). En otras palabras, *todas las muestras con $\alpha_i > 0$ satisfacen $y_i(\mathbf{w}\cdot \mathbf{x}_i + b)=1$*, definiendo el margen, mientras que aquellas bien dentro de su lado ($y_i(\mathbf{w}\cdot \mathbf{x}_i + b) > 1$) tienen $\alpha_i = 0$ y no influyen en $\mathbf{w}$.

Usando estos resultados, podemos formular el **problema dual** (maximizando $\mathcal{L}$ respecto a $\alpha_i$ tras sustituir $\mathbf{w}$ y $b$ óptimos). El problema dual para margen duro resulta ser:

$$
\begin{aligned}
\max_{\boldsymbol{\alpha}} \; & W(\boldsymbol{\alpha}) = \sum_{i=1}^n \alpha_i - \frac{1}{2}\sum_{i,j=1}^n \alpha_i\,\alpha_j\,y_i y_j\;\mathbf{x}_i\cdot \mathbf{x}_j, \\
\text{s.a. }\; & \sum_{i=1}^n \alpha_i y_i = 0, \qquad \alpha_i \ge 0 \ \forall i.
\end{aligned}
$$

Este es un problema de optimización cuadrática donde las **variables duales** son $\alpha_i$. La solución $\boldsymbol{\alpha}^*$ óptima nos da el vector $\mathbf{w}^*$ según la ecuación anterior, y podemos recuperar $b^*$ usando cualquier vector de soporte (por ejemplo, para un $j$ con $\alpha_j>0$, $b^* = y_j - \mathbf{w}^*\cdot \mathbf{x}_j$). En la versión con margen suave, el dual se modifica introduciendo cotas superiores $\alpha_i \le C$ (asociadas a las $\xi_i$), pero la idea general se mantiene: los $\alpha_i$ quedan acotados y solo un subconjunto serán no nulos (soporte).

La ventaja de resolver el problema dual es que puede ser más eficiente computacionalmente cuando $n$ (número de muestras) es menor que $p$ (dimensión), y además aparece de forma natural la posibilidad de reemplazar el producto interno $\mathbf{x}_i\cdot \mathbf{x}_j$ por una **función kernel** .

**Resumen de resultados importantes de la teoría SVM:**

- El hiperplano óptimo (en el caso lineal separable) está definido por $\mathbf{w}^* = \sum_{i \in SV} \alpha_i^* y_i \mathbf{x}_i$, una combinación de los vectores de soporte.
- Los **vectores de soporte** son aquellos para los cuales $\alpha_i^* > 0$. Geométricamente son las muestras que satisfacen $y_i(\mathbf{w}^*\cdot \mathbf{x}_i + b^*) = 1$, es decir, caen exactamente sobre el margen. Son los puntos más difíciles de clasificar y determinan el hiperplano.
- En problemas no separables, se introducen variables de holgura $\xi_i$ y un parámetro $C$ para equilibrar margen vs. error. **Cortes y Vapnik (1995)** demostraron que esta formulación de margen suave suele mejorar la generalización en problemas del mundo real.
- La solución del problema de optimización se puede obtener mediante algoritmos de QP o especializados. Un método notable es el algoritmo SMO (*Sequential Minimal Optimization*) de Platt, que resuelve el dual iterativamente seleccionando pares de $\alpha_i$ para optimizar, y es utilizado por muchas implementaciones eficientes de SVM.
- SVM no solo se utiliza para clasificación binaria lineal, sino que gracias al **truco del kernel** puede aplicarse a fronteras no lineales (transformando a un espacio de mayor dimensión donde el hiperplano corresponda a una frontera no lineal en el espacio original).

### Generación de un dataset sintético

Como caso de estudio, crearemos un conjunto de datos bidimensional sencillo con dos clases que sean aproximadamente separables linealmente. Por simplicidad, generaremos puntos distribuidos alrededor de dos centros distintos (por ejemplo, alrededor de $(-2,-2)$ y $(2,2)$) para las clases -1 y +1 respectivamente, agregando algo de ruido aleatorio. Esto nos permitirá visualizar el problema en 2D y probar el algoritmo.

````python
import numpy as np

# Fijar semilla para reproducibilidad
np.random.seed(0)

# Generar puntos para clase +1 (etiqueta = 1) alrededor de (2,2)
X_class1 = np.random.randn(50, 2) + np.array([2, 2])
y_class1 = np.ones(50)

# Generar puntos para clase -1 (etiqueta = -1) alrededor de (-2,-2)
X_class2 = np.random.randn(50, 2) + np.array([-2, -2])
y_class2 = -np.ones(50)

# Unir en un solo conjunto de datos
X = np.vstack([X_class1, X_class2])
y = np.concatenate([y_class1, y_class2])

print("Dimensiones de X:", X.shape)
print("Ejemplos de y:", y[:10], "...")
````

En el código anterior:
- Creamos 100 puntos en total (50 por clase) con dos características cada uno.
- `X_class1` son puntos de la clase positiva agrupados cerca de (2,2), y `X_class2` para la clase negativa cerca de (-2,-2).
- Combinamos ambos en los arrays `X` (100x2) e `y` (100,), donde `y` contiene 50 valores +1 seguidos de 50 valores -1.
- Imprimimos las dimensiones de `X` (que deberían ser (100,2)) y algunos ejemplos de `y` para verificar.

### Implementación paso a paso del algoritmo SVM desde cero

Para implementar SVM desde cero, podemos utilizar la formulación primal con margen suave. En lugar de resolver el problema de optimización mediante un paquete QP, implementaremos una solución iterativa basada en **descenso de gradiente subgradiente**. Usaremos la versión de minimización de la función de coste:

$$
L(\mathbf{w},b) = \frac{1}{2}\|\mathbf{w}\|^2 + C \sum_{i=1}^n \max(0,\,1 - y_i(\mathbf{w}\cdot \mathbf{x}_i + b)),
$$

donde la parte $\max(0,\,1 - y_i(\mathbf{x}_i\cdot\mathbf{w}+b))$ es la *pérdida hinge* que penaliza violaciones al margen. Este enfoque es equivalente a la optimización SVM y permite actualizar $\mathbf{w}$ gradualmente.

En cada iteración:
- Calculamos el gradiente (subgradiente) de la función de coste respecto a $\mathbf{w}$ y $b$.
- Si una muestra $i$ está **fuera o en el margen correcto** ($y_i(\mathbf{w}\cdot \mathbf{x}_i + b) \ge 1$), solo influye el término de regularización $\frac{1}{2}\|\mathbf{w}\|^2$ (el gradiente aporta $w$).
- Si una muestra $i$ **viola el margen** ($y_i(\mathbf{w}\cdot \mathbf{x}_i + b) < 1$), entonces además de la regularización debemos mover $\mathbf{w}$ y $b$ para reducir el error: esto implica sumar la contribución de $-C\,y_i \mathbf{x}_i$ al gradiente de $\mathbf{w}$ y de $-C\,y_i$ al gradiente de $b$ (que resulta en actualizar $b$ en dirección opuesta al signo de $y_i$).

Una implementación sencilla sigue este pseudocódigo:

```
Inicializar w = 0, b = 0
Para t = 1 hasta N_iter:
    Para cada muestra i:
        si y_i*(w · x_i + b) >= 1:
            w <- w - lr * (2*lambda * w)
            b <- b  (sin cambio, b no aparece en regularización)
        si no (violación):
            w <- w - lr * (2*lambda * w - y_i * x_i)
            b <- b - lr * (- y_i)   # que es b + lr * y_i, pero con signo cuidado
```

En el cual `lambda` es un parámetro de regularización (relacionado con $C$ inversamente) y `lr` es la tasa de aprendizaje. Básicamente, cuando la condición de margen se satisface, aplicamos solo la parte de decaimiento de pesos (regularización), y cuando no se satisface, ajustamos $\mathbf{w}$ y $b$ en dirección de corregir el error para la muestra $i$.

A continuación implementamos esto en una clase Python:

````python
class SVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate        # tasa de aprendizaje
        self.lambda_param = lambda_param  # parámetro de regularización (lambda)
        self.n_iters = n_iters        # número de iteraciones (epochs)
        self.w = None
        self.b = None

    def fit(self, X, y):
        # Inicializar parámetros
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0

        # Asegurar que y contiene valores en {-1, 1}
        y_ = np.where(y <= 0, -1, 1)  # convierte 0/False a -1, mantiene +1

        # Descenso subgradiente
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) + self.b) >= 1
                if condition:
                    # Caso 1: punto clasificado correctamente con margen -> solo regularización
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                    # (Nota: el sesgo b no se actualiza si no hay error,
                    # ya que no aparece en el término de regularización)
                else:
                    # Caso 2: violación del margen -> actualizar w y b
                    self.w -= self.lr * (2 * self.lambda_param * self.w - y_[idx] * x_i)
                    self.b -= self.lr * (- y_[idx])  # equivalente a self.b += self.lr * y_[idx]

    def predict(self, X):
        # Predicción basada en el signo de la función de decisión
        approx = np.dot(X, self.w) + self.b
        return np.sign(approx)
````


**Comentarios sobre la implementación:**

- La clase `SVM` encapsula el comportamiento del modelo. En `__init__` definimos la tasa de aprendizaje `lr`, el parámetro de regularización `lambda_param` (que corresponde conceptualmente a $ \lambda $ en la función de coste; un $ \lambda $ pequeño equivale a un $ C $ grande, dando más importancia a la minimización de errores), y `n_iters` el número de iteraciones (*epochs*) sobre los datos.

- En `fit(X, y)`, inicializamos $ \mathbf{w} $ como vector de ceros y $ b = 0 $. Luego convertimos el vector `y` a valores $-1$ y $+1$ (por si estuvieran etiquetados como 0/1).

- Iteramos `n_iters` veces sobre todo el *dataset* (bucle anidado). Para cada muestra, verificamos la condición
 $
    y_i \times (\mathbf{w}\cdot \mathbf{x}_i + b) \ge 1:
 $

  - Si se cumple, actualizamos $ \mathbf{w} $ con la deriva del término de regularización únicamente:
    
  $
      \mathbf{w} := \mathbf{w} - \text{lr} \times 2 \times \lambda \times \mathbf{w}.
  $
    (El factor 2 viene de derivar
  $
    \tfrac{1}{2}\|\mathbf{w}\|^2
  $
    , y `lambda_param` juega el rol de $ \lambda $ en ese término).

  - Si **no** se cumple la condición (es decir, el punto está mal clasificado o dentro del margen), actualizamos:
    - $
      \mathbf{w} := \mathbf{w} - \text{lr} \times \bigl( 2 \lambda\, \mathbf{w} - y_i\, \mathbf{x}_i \bigr).
     $
      Aquí $ -y_i\, \mathbf{x}_i $ es la derivada de $ C \, \max(0,\, 1 - y_i(\mathbf{w}\cdot\mathbf{x}_i+b)) $ respecto a $ \mathbf{w} $ (que es 0 si no hay error, y $ -y_i \mathbf{x}_i $ si hay error), y $ 2\lambda \mathbf{w} $ viene de la regularización.

      %
      b := b \;-\; \text{lr} \times (-y_i).
      %
      La derivada respecto a $ b $ de la *pérdida hinge* es $ -y_i $ (si hay violación), por lo que $ -(-y_i) = y_i $ se suma a $ b $. Nótese que en código usamos la forma equivalente `self.b -= lr * (-y_i)` para mantener consistencia de signos.

- El método `predict` simplemente calcula $ \mathbf{w}\cdot \mathbf{x} + b $ para cada muestra y devuelve el signo ($ +1 $ o $ -1 $).

Esta implementación es básica y puede requerir ajustes de hiperparámetros (`lr`, `lambda_param`, `n_iters`) según los datos para converger bien. Su complejidad es
$
O(\text{n_iters} \times n_\text{samples} \times p)
$
, que puede ser lenta para datos muy grandes, pero está bien para este ejemplo.

### Uso de la implementación y ejemplo de entrenamiento

Procedamos a entrenar nuestro modelo SVM en el *dataset* sintético generado y evaluar su rendimiento en entrenamiento:

````python
# Crear una instancia del SVM casero e entrenarla
svm_model = SVM(learning_rate=0.001, lambda_param=0.01, n_iters=1000)
svm_model.fit(X, y)

# Obtener predicciones en los datos de entrenamiento
y_pred = svm_model.predict(X)

# Calcular la precisión en entrenamiento
accuracy = np.mean(y_pred == np.where(y<=0, -1, 1))
print(f"Precisión en entrenamiento (modelo SVM propio): {accuracy*100:.2f}%")

# Imprimir los parámetros obtenidos
print("Vector de pesos w:", svm_model.w)
print("Sesgo b:", svm_model.b)
````

### Comparación con Scikit-Learn

Para validar nuestra implementación, vamos a entrenar un SVM usando la biblioteca scikit-learn en el mismo conjunto de datos y comparar resultados. Scikit-Learn utiliza internamente métodos optimizados (como la librería LIBSVM o similares) que resuelven el problema dual de forma eficiente. Usaremos un `SVC` (Support Vector Classifier) con kernel lineal. Ajustaremos el hiperparámetro `C` para que corresponda aproximadamente a nuestro $\lambda_param$ usado (por ejemplo, si $\lambda = 0.01$, eso sugiere $C \approx 1/\lambda = 100$ como valor razonable).

````python
from sklearn.svm import SVC

# Entrenar un SVM lineal con scikit-learn para comparar
svc_model = SVC(kernel='linear', C=100.0)  # C inversamente proporcional a lambda_param
svc_model.fit(X, y)

# Precisión en entrenamiento del modelo de scikit-learn
y_pred_svc = svc_model.predict(X)
accuracy_svc = np.mean(y_pred_svc == y)
print(f"Precisión en entrenamiento (Scikit-Learn SVC): {accuracy_svc*100:.2f}%")

# Parámetros del modelo scikit-learn
w_svc = svc_model.coef_[0]    # coeficientes del hiperplano
b_svc = svc_model.intercept_[0]
print("Vector de pesos w (Scikit-Learn):", w_svc)
print("Sesgo b (Scikit-Learn):", b_svc)

# Número de vectores de soporte y sus índices
print("Número de vectores de soporte (Scikit-Learn):", len(svc_model.support_))
print("Índices de vectores de soporte:", svc_model.support_)
````

El modelo de Scikit-Learn logra 100% de precisión en entrenamiento (posiblemente encontró una separación perfecta, lo cual sugiere que nuestro conjunto sintético era separable y que nuestro entrenamiento manual no llegó al óptimo global debido a parámetros de aprendizaje). Scikit-Learn identificó 2 vectores de soporte (sus índices en el array original se muestran, en este caso 16 y 99).

Observamos que los parámetros $w$ y $b$ difieren de los nuestros en escala, lo cual no es sorprendente. En SVM, el hiperplano puede representarse de múltiples formas escaladas. De hecho, la solución de Scikit-Learn probablemente esté **normalizada** de modo que los vectores de soporte satisfacen exactamente $y_i(w\cdot x_i + b) = 1$. Por eso, sus $w$ parecen más pequeños y $b$ distinto. Lo importante es que la frontera de decisión representada por $(w, b)$ es esencialmente la misma (podríamos comprobar que ambos hiperplanos clasifican de forma similar). También notamos que solo 2 puntos son vectores de soporte para el modelo entrenado por Scikit-Learn, lo que concuerda con la idea de que el hiperplano óptimo en un conjunto separable suele venir determinado por muy pocos puntos (los que están más cerca de la frontera).

En conclusión, nuestra implementación logra capturar el comportamiento del SVM y obtener un modelo razonable. Scikit-Learn, usando métodos más sofisticados, consigue el hiperplano óptimo exacto de margen máximo (con menos vectores de soporte y cero error en entrenamiento en este caso). Esto valida que los conceptos teóricos se reflejan en la práctica.

## 3. Visualización y Análisis

Ahora que tenemos modelos entrenados, podemos visualizar los **hiperplanos de separación** y los **márgenes** en nuestros datos sintéticos, así como analizar el efecto de distintos hiperparámetros. La visualización ayuda a construir la intuición de qué está haciendo el SVM.

### Representación gráfica del hiperplano de separación (2D)

Primero, grafiquemos el resultado en el plano 2D para nuestro dataset sintético. Usaremos el modelo de Scikit-Learn (`svc_model`) para obtener los vectores de soporte y trazar el hiperplano y las líneas de margen, ya que nos provee fácilmente estos elementos.

````python
import matplotlib.pyplot as plt

# Obtener los parámetros del modelo entrenado (Scikit-Learn) para graficar
w = svc_model.coef_[0]
b = svc_model.intercept_[0]
support_vectors = svc_model.support_vectors_

# Función auxiliar para trazar el hiperplano y márgenes
def plot_hyperplane(ax, w, b, style, label):
    # Calculamos dos puntos extremos en x para trazar la línea (en el rango de X)
    x0 = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100)
    # Para hiperplano: w0*x + w1*y + b = 0  ->  y = -(w0*x + b)/w1
    x1 = -(w[0]*x0 + b) / w[1]
    ax.plot(x0, x1, style, label=label)

# Crear figura
plt.figure(figsize=(6,5))
# Graficar las muestras, coloreando por clase
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o', label='Clase +1')
plt.scatter(X[y==-1, 0], X[y==-1, 1], color='red', marker='s', label='Clase -1')

# Graficar vectores de soporte con un contorno resaltado
plt.scatter(support_vectors[:,0], support_vectors[:,1], s=120,
            facecolors='none', edgecolors='k', linewidths=1.5, label='Vectores de soporte')

# Graficar el hiperplano y los márgenes
plot_hyperplane(plt, w, b, 'k-', label='Hiperplano')
# Márgenes: líneas donde w·x + b = ±1
plot_hyperplane(plt, w, b+1, 'k--', label='Márgenes')
plot_hyperplane(plt, w, b-1, 'k--', label=None)

plt.title('Hiperplano de Separación y Márgenes (SVM)')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend(loc='best')
plt.show()
````

En este gráfico:
- Los puntos azules y rojos representan las dos clases originales.
- Los círculos negros vacíos indican los **vectores de soporte** identificados.
- La línea negra sólida es el **hiperplano de separación** encontrado ($w\cdot x + b = 0$).
- Las líneas negras discontinuas representan los **límites del margen** ($w\cdot x + b = \pm 1$). Idealmente, los vectores de soporte deberían yacer sobre estas líneas.
- Podemos ver que el hiperplano (línea sólida) pasa equidistante de los vectores de soporte de ambas clases. Los vectores de soporte caen exactamente sobre las líneas de margen, lo que confirma que $y_i(w\cdot x_i + b) = 1$ para esos puntos.
- En nuestro caso, había 2 vectores de soporte, uno de cada clase, lo cual es típico en un problema linealmente separable sencillo: el hiperplano viene determinado por un par de puntos (en 2D) si el conjunto es estrictamente separable.

*(Al ejecutar el código en un entorno con capacidad gráfica, se visualizaría este diagrama.)*

### Hiperplano de separación en 3D

Para extender la intuición, consideremos datos en 3 dimensiones. En 3D, un hiperplano de separación es literalmente un **plano** que divide el espacio (por ejemplo, puede ser definido por una ecuación $w_0 x_0 + w_1 x_1 + w_2 x_2 + b = 0$). Generemos un dataset sintético 3D y veamos un ejemplo de SVM separando las clases:

````python
# Datos sintéticos 3D
np.random.seed(1)
# Clase +1 alrededor de (2,2,2)
X_class1_3d = np.random.randn(30,3) + np.array([2,2,2])
y_class1_3d = np.ones(30)
# Clase -1 alrededor de (-2,-2,-2)
X_class2_3d = np.random.randn(30,3) + np.array([-2,-2,-2])
y_class2_3d = -np.ones(30)
# Combinar
X3d = np.vstack([X_class1_3d, X_class2_3d])
y3d = np.concatenate([y_class1_3d, y_class2_3d])

# Entrenar SVM lineal en 3D
svc_3d = SVC(kernel='linear', C=100.0)
svc_3d.fit(X3d, y3d)
w3d = svc_3d.coef_[0]; b3d = svc_3d.intercept_[0]
print("Hiperplano 3D: {} · x + ({}) = 0".format(w3d, b3d))
print("Número de vectores de soporte (3D):", len(svc_3d.support_))
````

Podemos graficar este plano en 3D junto con los puntos (aunque la visualización en 3D es más complicada en un medio estático):

````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')
# Graficar puntos de ambas clases
ax.scatter(X3d[y3d==1,0], X3d[y3d==1,1], X3d[y3d==1,2], c='blue', marker='o', label='Clase +1')
ax.scatter(X3d[y3d==-1,0], X3d[y3d==-1,1], X3d[y3d==-1,2], c='red', marker='^', label='Clase -1')
# Graficar vectores de soporte
support_3d = svc_3d.support_vectors_
ax.scatter(support_3d[:,0], support_3d[:,1], support_3d[:,2], s=100, facecolors='none', edgecolors='k', label='Vectores de soporte')

# Crear una malla para dibujar el plano:
xx, yy = np.meshgrid(np.linspace(X3d[:,0].min()-1, X3d[:,0].max()+1, 10),
                     np.linspace(X3d[:,1].min()-1, X3d[:,1].max()+1, 10))
# Calcular z = -(w0*x + w1*y + b)/w2
zz = -(w3d[0]*xx + w3d[1]*yy + b3d) / w3d[2]
# Graficar la superficie del hiperplano
ax.plot_surface(xx, yy, zz, alpha=0.3, color='green', label='Hiperplano')

ax.set_title('Hiperplano de Separación en 3D')
ax.set_xlabel('X1'); ax.set_ylabel('X2'); ax.set_zlabel('X3')
plt.legend(loc='upper right')
plt.show()
````

En esta visualización 3D:
- Los puntos azules y rojos son las muestras de cada clase.
- Los círculos negros indican vectores de soporte.
- El plano semitransparente verde es el hiperplano $w\cdot x + b = 0$ hallado.
- Podemos verificar que separa correctamente la mayoría de puntos azules arriba de rojos abajo (o viceversa según orientación). Los vectores de soporte estarán sobre ese plano o muy cerca de él en el borde.

*(De nuevo, en un entorno interactivo se podría rotar la vista para apreciar mejor la separación.)*

Esta representación confirma que la idea del hiperplano separador se extiende a dimensiones superiores: en 3D es un plano que deja los datos de distintas clases a cada lado.

### Impacto de distintos hiperparámetros en la clasificación

Finalmente, analicemos el efecto del hiperparámetro $C$ (margen vs error) en la solución de SVM. Usaremos un dataset donde las clases **no son perfectamente separables** (introduciremos mayor solapamiento) para ver cómo un $C$ grande vs $C$ pequeño producen fronteras diferentes.

Generemos un conjunto con más traslape:


````python
np.random.seed(2)
# Clase +1 alrededor de (1,1), Clase -1 alrededor de (-1,-1)
X_plus_overlap = np.random.randn(50,2) + np.array([1.0, 1.0])
X_minus_overlap = np.random.randn(50,2) + np.array([-1.0, -1.0])
y_plus_overlap = np.ones(50)
y_minus_overlap = -np.ones(50)
X_overlap = np.vstack([X_plus_overlap, X_minus_overlap])
y_overlap = np.concatenate([y_plus_overlap, y_minus_overlap])

# Entrenar SVM con C alto (casi hard margin) y C bajo (soft margin)
svc_highC = SVC(kernel='linear', C=1000.0).fit(X_overlap, y_overlap)
svc_lowC  = SVC(kernel='linear', C=0.1).fit(X_overlap, y_overlap)

print("Support vectors (C alto):", len(svc_highC.support_))
print("Support vectors (C bajo):", len(svc_lowC.support_))
print("Precisión entrenamiento (C alto):", np.mean(svc_highC.predict(X_overlap)==y_overlap))
print("Precisión entrenamiento (C bajo):", np.mean(svc_lowC.predict(X_overlap)==y_overlap))
````

Interpretación:
- Con $C$ muy alto, el modelo intenta clasificar **perfectamente** o con mínimo error, resultando en 92% de precisión (aún así, no llega a 100% porque los datos se solapan, pero trata de reducir errores al mínimo). Se obtuvieron 26 vectores de soporte.
- Con $C$ bajo, el modelo permite más errores para lograr un margen más amplio. Obtuvo 89% de precisión (permitió algunos errores más) y tiene **más vectores de soporte (39)**. Esto concuerda con la idea de que un margen más amplio (menos restricciones) suele involucrar más puntos cerca de la frontera, mientras que un modelo que intenta ajustar más estrictamente tiene menos vectores de soporte *críticos* y algunos puntos quedan correctamente clasificados con amplio margen.

Visualicemos ambas situaciones:

````python
# Parámetros de cada modelo
w_high, b_high = svc_highC.coef_[0], svc_highC.intercept_[0]
w_low, b_low   = svc_lowC.coef_[0], svc_lowC.intercept_[0]

plt.figure(figsize=(10,4))
# Subplot 1: C alto
plt.subplot(1,2,1)
plt.scatter(X_overlap[y_overlap==1,0], X_overlap[y_overlap==1,1], color='blue', label='Clase +1')
plt.scatter(X_overlap[y_overlap==-1,0], X_overlap[y_overlap==-1,1], color='red', label='Clase -1')
plot_hyperplane(plt, w_high, b_high, 'k-', label='Hiperplano (C alto)')
plot_hyperplane(plt, w_high, b_high+1, 'k--', label='Márgenes')
plot_hyperplane(plt, w_high, b_high-1, 'k--', label=None)
plt.title(f'SVM con C alto (={svc_highC.C})')
plt.legend(loc='upper right')

# Subplot 2: C bajo
plt.subplot(1,2,2)
plt.scatter(X_overlap[y_overlap==1,0], X_overlap[y_overlap==1,1], color='blue', label='Clase +1')
plt.scatter(X_overlap[y_overlap==-1,0], X_overlap[y_overlap==-1,1], color='red', label='Clase -1')
plot_hyperplane(plt, w_low, b_low, 'g-', label='Hiperplano (C bajo)')
plot_hyperplane(plt, w_low, b_low+1, 'g--', label='Márgenes')
plot_hyperplane(plt, w_low, b_low-1, 'g--', label=None)
plt.title(f'SVM con C bajo (={svc_lowC.C})')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

En la figura comparativa:
- Con **C alto** (izquierda, líneas negras), el hiperplano se orienta de manera a minimizar errores: podemos ver que intenta separar incluso puntos difíciles, resultando en un margen más estrecho (las líneas punteadas de margen están más cerca del hiperplano).
- Con **C bajo** (derecha, líneas verdes), el hiperplano es diferente; parece no esforzarse en clasificar algunos puntos conflictivos, permitiendo que queden del lado contrario, a cambio de conseguir un margen más amplio entre las clases principales. Las líneas de margen verdes están más separadas del hiperplano verde. Este modelo ignora ciertos puntos ruidosos (posiblemente considerándolos casi outliers), lo que podría mejorar la generalización.
- En resumen, un **C bajo** produce un clasificador más *simple* (margen amplio, más regularización, algunos errores en entrenamiento) mientras que un **C alto** produce un clasificador más *complejo* (margen estrecho, menos errores en entrenamiento pero potencialmente más sobreajuste). Esto se corresponde con la teoría del parámetro $C$ en SVM.

*(Las gráficas confirmarían visualmente esta diferencia: el hiperplano verde (C bajo) estaría más centrado y suave, mientras que el negro (C alto) se torcería más hacia los puntos difíciles.)*

**Conclusión de la sección de análisis:** Hemos visto cómo el SVM encuentra hiperplanos que maximizan el margen en 2D y 3D, y cómo los vectores de soporte determinan ese hiperplano. Además, experimentamos con el parámetro $C$, observando que afecta el equilibrio entre margen amplio vs. error en entrenamiento. Estas propiedades hacen de SVM un clasificador poderoso: trata de lograr la mayor separación posible entre clases, pero también puede flexibilizarse para no sobreajustar en datos ruidosos.
